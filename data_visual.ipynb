```python
import pandas as pd
```

```python
df = pd.read_csv('Covid19 Harian Indonesia.csv',sep=',')
df.head(10)
```

```python
import seaborn as sns

sns.pairplot(df)
```

```python
corr = df.corr()
sns.heatmap(corr)
```

```python
sns.heatmap(corr, cmap=['red','green','blue'],
vmin = -.5 , vmax = 0.6,annot = True)
```

Dari hasil korelasi diatas dapat diketahui bahwa beberapa variabel memiliki korelasi yang kuat diantaranya:
1. Jumlah meninggal dengan jumlah sembuh
2. Jumlah positif dengan jumlah dirawat kumulatif
3. Jumlah dirawat dengan jumlah positif
4. Jumlah positif kumulatif dengan jumlah meninggal kumulatif
5. Jumlah sembuh kumulatif dengan jumlah sembuh
6. Jumlah dirawat kumulatif dengan jumlah sembuh
